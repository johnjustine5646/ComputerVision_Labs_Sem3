```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Read the original image and add noise
original_img = cv2.imread('/content/birdimg.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Add noise to the image (for example, Gaussian noise)
mean = 0
variance = 100
sigma = np.sqrt(variance)
gaussian_noise = np.random.normal(mean, sigma, original_img.shape)
noisy_image = original_img + gaussian_noise.astype(np.uint8)
```

```python
# Step 2: Apply filtering techniques to restore the image
# Filtering techniques: median filter, Gaussian filter, bilateral filter, etc.

# Median filter
median_filtered = cv2.medianBlur(noisy_image, 5)  # Adjust kernel size as needed

# Gaussian filter
gaussian_filtered = cv2.GaussianBlur(noisy_image, (5, 5), 0)  # Adjust kernel size as needed

# Bilateral filter
bilateral_filtered = cv2.bilateralFilter(noisy_image, 9, 75, 75)  # Adjust parameters as needed
```

```python
# Display the results
plt.figure(figsize=(12, 8))

plt.subplot(231)
plt.title('Original Image with Noise')
plt.imshow(noisy_image, cmap='gray')
plt.axis('off')

plt.subplot(232)
plt.title('Median Filter')
plt.imshow(median_filtered, cmap='gray')
plt.axis('off')

plt.subplot(233)
plt.title('Gaussian Filter')
plt.imshow(gaussian_filtered, cmap='gray')
plt.axis('off')

plt.subplot(234)
plt.title('Bilateral Filter')
plt.imshow(bilateral_filtered, cmap='gray')
plt.axis('off')

plt.show()
```
