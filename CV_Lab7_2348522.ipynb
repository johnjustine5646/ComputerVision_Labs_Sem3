```python
import cv2
import numpy as np
```

```python
# Step 1: Read the original image and add noise
img_path = "/content/download.jpg"  # Change this to the path of your original image
img = cv2.imread(img_path)
```

```python
# Adding noise to the image
noise = np.random.normal(loc=0, scale=50, size=img.shape)
noisy_img = np.clip(img + noise, 0, 255).astype(np.uint8)
```

```python
# Step 2: Apply filtering techniques to restore the image
# Here, we'll use Gaussian blur as a simple filtering technique
restored_img = cv2.GaussianBlur(noisy_img, (5, 5), 0)
```

```python
from google.colab.patches import cv2_imshow
# Display original, noisy, and restored images
print('Original Image')
cv2_imshow(img)
print('Noisy Image')
cv2_imshow(noisy_img)
print('Restored Image')
cv2_imshow(restored_img)
```
