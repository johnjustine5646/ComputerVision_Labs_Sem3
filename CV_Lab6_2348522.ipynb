```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Read the image
img = cv2.imread('/content/photoshop1Kelly.jpg')
```

```python
# Median filter
median = cv2.medianBlur(img, 5)  # Change the kernel size as needed
```

```python
# Max and min filters
max_filter = cv2.dilate(img, None, iterations=1)
min_filter = cv2.erode(img, None, iterations=1)
```

```python
# Mid-point filter
max_filter = cv2.dilate(img, None, iterations=1)
min_filter = cv2.erode(img, None, iterations=1)
mid_point = (max_filter + min_filter) / 2
```

```python
# Alpha-trimmed filter
def alpha_trimmed(img, d):
    result = np.zeros_like(img)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            pixels = []
            for k in range(-d//2, d//2 + 1):
                for l in range(-d//2, d//2 + 1):
                    if 0 <= i + k < img.shape[0] and 0 <= j + l < img.shape[1]:
                        pixels.append(img[i+k, j+l])
            pixels.sort()
            pixels = pixels[d//2:-d//2]
            result[i, j] = int(sum(pixels) // len(pixels)) # Convert to int for compatibility
    return result

alpha_trimmed_filter = alpha_trimmed(img, d=3)  # Change the parameter d as needed
```

```python
# Displaying the results using matplotlib
plt.figure(figsize=(12, 10))

plt.subplot(231)
plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(232)
plt.title('Median Filter')
plt.imshow(median, cmap='gray')
plt.axis('off')

plt.subplot(233)
plt.title('Max Filter')
plt.imshow(max_filter, cmap='gray')
plt.axis('off')

plt.subplot(234)
plt.title('Min Filter')
plt.imshow(min_filter, cmap='gray')
plt.axis('off')

plt.subplot(235)
plt.title('Mid-Point Filter')
plt.imshow(mid_point.astype(np.uint8), cmap='gray')
plt.axis('off')

plt.subplot(236)
plt.title('Alpha-Trimmed Filter')
plt.imshow(alpha_trimmed_filter.astype(np.uint8), cmap='gray')
plt.axis('off')

plt.show()
```
