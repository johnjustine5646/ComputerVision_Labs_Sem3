```python
#Extra Prg-1 Illustrate the following image Manipulation operation
#i)	Read, write and view the images using different python libraries
# (CV2, ImageIO, MatplotLib, PIL)
#ii)	Convert the given images in to different formats
# (For ex: *.jpeg, *.jpg, *.jpe, , *.png, tiff, *.tif, e.t.c)
```

```python
# Importing required libraries
import cv2
import imageio
import matplotlib.pyplot as plt
from PIL import Image

# i) Read, write, and view images using different libraries

# Read an image using OpenCV (cv2)
image_cv2 = cv2.imread('/content/tiger.jpg')

# Write an image using OpenCV (cv2)
cv2.imwrite('output.jpg', image_cv2)

# Display the image using Matplotlib
plt.imshow(image_cv2)
plt.title('Image using Matplotlib')
plt.show()
```

```python
# Read an image using ImageIO
image_imageio = imageio.imread('/content/tiger.jpg')

# Write an image using ImageIO
imageio.imwrite('output_image_imageio.jpg', image_imageio)

# Display the image using Matplotlib
plt.imshow(image_imageio)
plt.title('Image using Matplotlib')
plt.show()
```

```python
# Read an image using PIL
image_pil = Image.open('/content/tiger.jpg')

# Write an image using PIL
image_pil.save('output.jpg')

# Display the image using Matplotlib
plt.imshow(image_pil)
plt.title('Image using Matplotlib')
plt.show()
```

```python
# Convert image using OpenCV (cv2)
cv2.imwrite('output1.png', image_cv2)
# Convert image using ImageIO
imageio.imwrite('output2.png', image_imageio)
# Convert image using PIL
image_pil.save('output3.png')
# Convert image format using PIL
image_pil.save('output4.tiff', format='TIFF')
```
