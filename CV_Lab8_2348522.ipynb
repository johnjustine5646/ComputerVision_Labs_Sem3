```python
import cv2
import numpy as np

# Read the image
img_path = "/content/images.jpeg"  # Change this to the path of your original image
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

# First Order Derivative Methods (Sobel)
sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
sobel_combined = cv2.magnitude(sobel_x, sobel_y)

# Second Order Derivative Methods (Laplacian)
laplacian = cv2.Laplacian(img, cv2.CV_64F)

# Optimum Edge Detection Method (Canny)
canny = cv2.Canny(img, 100, 200)
from google.colab.patches import cv2_imshow

# Display original and edge-detected images
print('Original Image')
cv2_imshow(img)
print('Sobel Edge Detection')
cv2_imshow( sobel_combined.astype(np.uint8))
print('Laplacian filetr')
cv2_imshow( laplacian.astype(np.uint8))
print('Canny Edge Detection')
cv2_imshow( canny)
```
