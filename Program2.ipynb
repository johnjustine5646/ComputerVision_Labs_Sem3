```python
#Libraries
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
from scipy.signal import convolve2d
```

```python
img = cv2.imread('/content/download.jpg')
img
```

```python
kernel_identity = np.array([[0, 0, 0],
                                [0, 1, 0],
                                [0, 0, 0]])




identity_result = cv2.filter2D(img, -1, kernel_identity)

cv2_imshow(img)
cv2_imshow(identity_result)
```

```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
from scipy.signal import convolve2d

img = cv2.imread('/content/manhattanhenge1.jpeg', 0)

identity_kernel = np.array([[0, 0, 0],
                            [0, 1, 0],
                            [0, 0, 0]])

identity_result = convolve2d(img, identity_kernel, mode='same', boundary='fill', fillvalue=0)

cv2_imshow(img)
cv2_imshow(identity_result.astype('uint8'))
```

```python
edge_kernel = np.array([[-1, -1, -1],
                        [-1,  8, -1],
                        [-1, -1, -1]])


sharpen_kernel = np.array([[0, -1, 0],
                           [-1, 5, -1],
                           [0, -1, 0]])

box_blur_kernel = np.array([[1, 1, 1],
                            [1, 1, 1],
                            [1, 1, 1]]) / 9

gaussian_blur_kernel = np.array([[1, 2, 1],
                                 [2, 4, 2],
                                 [1, 2, 1]]) / 16

edge_result = cv2.filter2D(img, -1, edge_kernel)


sharpen_result = cv2.filter2D(img, -1, sharpen_kernel)


box_blur_result = convolve2d(img, box_blur_kernel, mode='same', boundary='fill', fillvalue=0)


gaussian_blur_result = convolve2d(img, gaussian_blur_kernel, mode='same', boundary='fill', fillvalue=0)
```

```python
cv2_imshow(img)
cv2_imshow(edge_result)
cv2_imshow(sharpen_result)
cv2_imshow(box_blur_result.astype('uint8'))
cv2_imshow(gaussian_blur_result.astype('uint8'))
```

```python
modes = ['valid', 'same', 'full']

for mode in modes:
    conv_result = convolve2d(img, identity_kernel, mode=mode, boundary='fill', fillvalue=0)
    cv2_imshow(conv_result.astype('uint8'))
```

```python
img = cv2.imread('/content/manhattanhenge1.jpeg',0)


modes = ['reflect', 'constant', 'nearest', 'mirror', 'wrap']
for mode in modes:
    print(mode)
    conv_result = convolve(img, identity_kernel, mode=mode)
    cv2_imshow(conv_result.astype('uint8'))
```
